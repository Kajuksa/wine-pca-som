## 1. Introduction to the problem of study [1,0 valor]
**Wine quality assessment** is a critical task in the wine industry, traditionally performed through sensory evaluation by experts. However, this process is subjective, time-consuming, and costly. Considering the availability of physicochemical data (such as acidity, alcohol content, sulphates, and pH), there is an opportunity to apply data-driven techniques to analyze and predict the wine quality objectively.

The main challenge lies in the high dimensionality and correlation among features: many chemical properties are interrelated, which can complicate analysis and modeling.

To address this, **Principal Component Analysis (PCA)** is implemented as a dimensionality reduction technique. PCA transforms correlated variables into a smaller set of uncorrelated components, while preserving most of the variance present in the data. This simplifies the dataset, enhances visualization, and improves the efficiency of predictive models.

Besides PCA, this project also applies **Self-Organizing Maps (SOM)**, a neural-based dimensionality reducion and clustering method (introduced by Kohonen). SOM provides a non-linear, topology-preserving projection of the wine data into a 2-dimensionally grid, allowing the:
- visualization of the neighbourhood relations between wines;
- detection of clusters;
- investigation of how chemical descriptors map to different regions;
- comparison with PCA's linear projections.

**Together**, PCA and SOM, provide two complementary types of views:
- PCA: linear, variance-preserving projection;
- SOM: non-linear, topology-preserving mapping.

Using both, allows a better understanding of the structure in the wine dataset.

## 3. Brief History and literature review of the problem and methods/algorithms [1,5 valor]

### **Principal Component Analysis (PCA):**

**Principal Component Analysis (PCA)** was first introduced by **Karl Pearson** in 1901 as a method for reducing the dimensionality of data while preserving its essential structure. Later, Harold Hotelling (1933) formalized PCA in the context of multivariate statistics, making it a cornerstone technique in data analysis.
Over the decades, PCA has become a fundamental tool in statistics, machine learning, and signal processing, widely applied in fields such as image compression, genomics, finance, and pattern recognition. Its popularity stems from its ability to simplify complex datasets by projecting them onto a smaller set of orthogonal components that capture the maximum variance.

- **Pearson (1901)**: Introduced PCA as a least-squares approach to fitting lines and planes to data.
- **Hotelling (1933)**: Extended PCA to multivariate analysis, introducing eigenvalue decomposition.
- **Jolliffe (1986)**: Published Principal Component Analysis, a comprehensive reference that remains influential.

Modern Extensions:

- **Kernel PCA**: Introduced by Schölkopf et al. (1998) for nonlinear dimensionality reduction using kernel methods;
- **Sparse PCA**: Developed to improve interpretability by enforcing sparsity in components;
- **Probabilistic PCA**: Introduced by Tipping and Bishop (1999) to model PCA in a probabilistic framework.

### **Self-Organizing Maps (SOM):**

**Self-Organizing Maps (SOM):** were introduced by Teuvo Kohonen in 1982, and became one of the most influential unsupervised neural algorithms.

Key contributions:
- **Kohonen (1982):** Proposed SOM as a model of biological topographic maps;
- **Kohonen (1995):** Published Self-Organizing Maps.

SOM are used in:
- exploratory data analysis;
- clustering;
- visualization;
- anomaly detection;
- chemometrics (wine analysis, sensory evaluation, ...).

SOM is useful when PCA's linear assumptions are too restrictive. SOM maps non-linear manifolds into low-dimensional topology surfaces.

## 4. About the main method/algorithm used [1,5 valor]

### **Principal Component Analysis (PCA):**

Principal Component Analysis (PCA) is a widely used dimensionality reduction technique in data analysis and machine learning. It transforms a high-dimensional dataset into a lower-dimensional space, while preserving as much variability (information) as possible. PCA achieves this by identifying new axes, called principal components, which are linear combinations of the original features and capture the directions of maximum variance in the data.

The main goals of PCA are:

- **Reduce complexity**: By lowering the number of features, PCA simplifies models and speeds up computations.
- **Remove redundancy**: It eliminates correlated features by projecting data onto orthogonal components.
- **Visualize high-dimensional data**: PCA enables plotting in 2D or 3D for better interpretation.

Principal Component Analysis (PCA) is an unsupervised linear transformation technique used for dimensionality reduction and feature extraction. The core idea is to project high-dimensional data onto a new set of orthogonal axes (principal components) that capture the directions of maximum variance in the dataset.

The approach of PCA is based on reducing the dimensionality of a dataset while retaining the most significant information. It identifies new variables, called principal components, which are linear combinations of the original features and capture the directions of maximum variance.

**Steps Involved:**

- **Data Standardization**:
Normalize the dataset so that all features contribute equally, especially when they have different scales.

- **Covariance Matrix Calculation**:
Compute the covariance matrix to understand how variables relate to each other.

- **Eigen Decomposition**:
Find eigenvalues and eigenvectors of the covariance matrix.
Eigenvectors define the directions of the new axes (principal components).
Eigenvalues indicate the amount of variance captured by each component.

- **Component Selection**:
Rank components by their eigenvalues and select the top components that explain most of the variance.

- **Projection**:
Transform the original data onto the selected principal components to obtain a reduced representation.

### **Self-Organizing Maps (SOM):**

Self-Organizing Maps are unsupervised neural networks used for non-linear dimensionality reduction, clustering, and data visualization, while preserving the topological relations. Unlike PCA, which is linear, SOM projects the data into a 2D grid of neurons, allowing an intuitive visual interpretation of high-dimensional patterns.

Each neuron in the map holds a weight vector $w_k$, with the same dimension as the input samples. During the training, the neurons compete, and only 1 neuron (the **Best Matching Unit (BMU)**) "wins", and him and its neighbors are updated.

- **Distance to determine the BMU:**
The neuron whose weight vector is closest to the input $x(t)$ os selected as the "winner":
$$
d_k(t) = \lVert x(t) - w_k(t) \rVert^2
$$
$$
d_c(t) = min_k(d_k(t))
$$

- **Weight update rule:**
$$
w_k(t + 1) = w_k(t) + α(t)[x(t) - w_k(t)] \text{,}
$$
Where:
  - $α(t)$ is the learning rate, decreasing over the time.

- **Neighborhood function:**
BMU influences nearby neurons using a Gaussian neighborhood function:
$$
h_{ck}(t) = exp(\frac{\lVert r_k - r_c \rVert^2}{θ(t)^2})
$$
Where:
  - $r_k, r_c$ = coordinates of neurons $k$ and $c$ on the 2D grid;
  - $θ(t)$ = neighborhood radius (shrinks during training).

**Why is SOM useful here:**
- Captures non-linear structures, missed by PCA;
- Produces visual topology maps of wine characteristics;
- Reduces dimensionality, while preserving the neighborhood structure.

## 7. Main code as possible solution to the problem [1,5 valor]

```python
# Standardize features (mean=0, variance=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_dedup)

print("Standardized matrix shape:", X_scaled.shape)
```

### **7.1. PCA:**

```python
# Fit PCA with all components
pca = PCA(random_state=RANDOM_STATE)
X_pca = pca.fit_transform(X_scaled)

explained_ratio = pca.explained_variance_ratio_
cum_explained = np.cumsum(explained_ratio)

print("Explained variance ratio by component:")
for i, r in enumerate(explained_ratio, start=1):
    print(f"  PC{i}: {r:.4f}")

print("\nCumulative explained variance:")
for i, r in enumerate(cum_explained, start=1):
    print(f"  up to PC{i}: {r:.4f}")
```

```python
# Choose k so that cumulative explained variance >= 95%
k95 = int(np.argmax(cum_explained >= 0.95)) + 1  # +1 because argmax gives index
print(f"\nSelected k (>=95% variance): {k95}")
```

```python
# Fit PCA with k95 components
pca_k = PCA(n_components=k95, random_state=RANDOM_STATE)
X_pca_k = pca_k.fit_transform(X_scaled)

# Create loadings and scores DataFrames
loadings_all = pd.DataFrame(
    pca.components_.T, index=feature_cols,
    columns=[f'PC{i+1}' for i in range(len(feature_cols))]
)
scores_all = pd.DataFrame(
    X_pca, columns=[f'PC{i+1}' for i in range(len(feature_cols))]
)

loadings_k = pd.DataFrame(
    pca_k.components_.T, index=feature_cols,
    columns=[f'PC{i+1}' for i in range(k95)]
)
scores_k = pd.DataFrame(
    X_pca_k, columns=[f'PC{i+1}' for i in range(k95)]
)

display(loadings_k.head())
display(scores_k.head())
```

```python
# Scree plot + cumulative variance
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scree plot
axes[0].bar(range(1, len(explained_ratio) + 1), explained_ratio, color='#4C72B0')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance Ratio')
axes[0].set_title('Scree Plot')

# Cumulative explained variance
axes[1].plot(range(1, len(cum_explained) + 1), cum_explained, marker='o', color='#55A868')
axes[1].axhline(0.95, color='red', linestyle='--', label='95% threshold')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Explained Variance')
axes[1].set_title('Cumulative Explained Variance')
axes[1].legend()

# Save figures
fig.savefig('pca_scree_cumulative.png', dpi=150)

plt.tight_layout()
plt.show()

```

```python
# Biplot for PC1–PC2
fig, ax = plt.subplots(figsize=(7, 6))
ax.scatter(scores_all['PC1'], scores_all['PC2'], alpha=0.6, s=16)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('PCA Biplot (scores and loadings)')

# Loadings arrows
scale_vec = 3  # increase for longer arrows
for i, v in enumerate(loadings_all[['PC1', 'PC2']].values):
    ax.arrow(0, 0, v[0] * scale_vec, v[1] * scale_vec,
             color='r', alpha=0.5, head_width=0.05)
    ax.text(v[0] * scale_vec * 1.1, v[1] * scale_vec * 1.1,
            loadings_all.index[i], color='r', fontsize=9)

# Save
fig.savefig('pca_biplot.png', dpi=150)

plt.tight_layout()
plt.show()

```

```python
# Loadings heatmap (absolute values)
fig = plt.figure(figsize=(8, max(4, len(feature_cols) * 0.3)))
sns.heatmap(np.abs(loadings_all), annot=False, cmap='viridis')
plt.title('Absolute Loadings of Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Features')

# Save
fig.savefig('pca_loadings_heatmap.png', dpi=150)

plt.tight_layout()
plt.show()

```

### **7.2. SOM:**

```python
# SOM grid size
som_x, som_y = 15, 15

# Initialize the SOM
som = MiniSom(
    x = som_x,
    y = som_y,
    input_len = X_scaled.shape[1],
    sigma = 1.2,
    learning_rate = 0.5,
    neighborhood_function = "gaussian",
    random_seed = RANDOM_STATE
)

som.random_weights_init(X_scaled)
```

```python
# Train SOM
print("Training SOM...")
som.train_random(data=X_scaled, num_iteration=3000)
print("SOM training completed!")
```

```python
# Extract the BMU coordinates for each sample (used as features)
som_coords = np.array([som.winner(x) for x in X_scaled])
som_df = pd.DataFrame(som_coords, columns=["SOM_x", "SOM_y"])

display(som_df.head())
```

```python
# U-Matrix (distance map)
u_matrix = som.distance_map()

plt.figure(figsize=(7, 7))
plt.imshow(u_matrix, cmap="bone")
plt.title("SOM U-Matrix (Distance Map)")
plt.colorbar()
plt.savefig("som_umatrix.png", dpi=150)
plt.tight_layout()
plt.show()
```

```python
# Component (feature) maps
for i, feature in enumerate(feature_cols):
  plt.figure(figsize=(6, 6))
  plt.title(f"SOM Component Map - {feature}")
  plt.imshow(som.get_weights()[:, :, i], cmap="viridis")
  plt.colorbar()
  plt.savefig(f"som_component_{feature}.png", dpi=150)
  plt.tight_layout()
  plt.show()
```

```python
# Adding SOM coordinates to the dataframe for later analysis
df_som_features = som_df.copy()
```

```python
# Quality mapping on the SOM
# 15x15 matrix , where each neuron stores the average wine quality
quality_map = np.zeros((som_x, som_y))

# Count how many samples fall in each neuron
count = np.zeros((som_x, som_y))

for idx, x in enumerate(X_scaled):
  bx, by = som.winner(x)
  quality_map[bx, by] += df.loc[X_dedup.index[idx], "quality"]
  count[bx, by] += 1

quality_map = np.divide(quality_map, count, out=np.zeros_like(quality_map), where=count!=0)

# Plot of the quality map
fig, ax = plt.subplots(figsize=(7, 7))
im = ax.imshow(quality_map, cmap="viridis")
plt.title("SOM Quality Map (Average quality per neuron)")
plt.colorbar(im)
plt.tight_layout()
fig.savefig("som_quality_map.png", dpi=150)
plt.show()
```

- Input: same standardized 11 physicochemical variables used in PCA

- Goal: project high-dimensional wine data onto a 2D topology-preserving map

**2) U-Matrix (Distance Map)**

- Bright/white areas → large distances between neighboring neurons → **cluster boundaries**
- Dark areas → similar weight vectors → **homogeneous groups of wines**

- Two strong bright zones clearly separate the main clusters
- Large dark patches indicate compact, chemically similar wine groups
- The overall structure is highly **granular and curved** compared to PCA → confirming detection of **non-linear manifolds**

These maps clearly reveal which chemical properties drive the topological organization of the data.

**4) Quality mapping on the SOM**

When wine quality is projected onto the trained SOM:
- High-quality wines (6–7) cluster tightly in **low volatile acidity / high alcohol** regions
- Low-quality wines (3–4) concentrate around neurons with **high SO₂ and high volatile acidity**

Our quality map uses a purple-to-yellow scale:
- Yellow/bright zones = higher average quality
- Purple/dark zones = lower average quality

Key insights:
- High-quality wines form compact clusters, especially in the **upper-middle and middle-left** areas — perfectly matching high-alcohol / low-volatile-acidity zones
- Low-quality wines appear in several distinct pockets (bottom-left, mid-right, and high-SO₂ regions)
- **Quality boundaries closely follow U-Matrix bright lines** → quality is not random but deeply embedded in the learned chemical topology

**SOM conclusions**

The SOM fully confirms PCA findings while adding rich non-linear detail, revealing natural clustering patterns and providing highly intuitive 2D visualizations (U-matrix, component planes, and quality map) that greatly enhance the understanding of wine chemistry and quality relationships.

**UMAP + HDBSCAN represents the current gold standard in exploratory data analysis and delivers the clearest, most intuitive visualization of the hidden structure in the WineQT dataset.**

```python
import umap
import hdbscan

df = pd.read_csv('WineQT.csv')
df = df.drop(columns=['Id'], errors='ignore')

print(f"Dataset loaded → {df.shape[0]} samples, {df.shape[1]} columns")
print("Columns:", df.columns.tolist())

X = df.drop(columns=['quality'])        # 11 physicochemical features
y = df['quality']                        # target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"X_scaled ready → shape: {X_scaled.shape}")

# 3. UMAP projection
reducer = umap.UMAP(
    n_neighbors=30,
    min_dist=0.0,
    n_components=2,
    metric='euclidean',
    random_state=42
)
umap_embedding = reducer.fit_transform(X_scaled)

# 4. HDBSCAN clustering
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=35,
    min_samples=15,
    cluster_selection_epsilon=0.5,
    cluster_selection_method='eom'
)
cluster_labels = clusterer.fit_predict(umap_embedding)

plt.figure(figsize=(12, 9))
scatter = plt.scatter(umap_embedding[:, 0], umap_embedding[:, 1],
                     c=cluster_labels, cmap='Spectral', s=50,
                     edgecolor='k', linewidth=0.3, alpha=0.85)
plt.colorbar(scatter, label='Cluster ID (-1 = outlier)')
plt.title('WineQT – UMAP + HDBSCAN\nAutomatic Clusters Detected', fontsize=16, pad=20)
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(12, 9))
scatter2 = plt.scatter(umap_embedding[:, 0], umap_embedding[:, 1],
                      c=y, cmap='viridis', s=50,
                      edgecolor='k', linewidth=0.3, alpha=0.85)
plt.colorbar(scatter2, label='Wine Quality (3–8)')
plt.title('WineQT – UMAP Projection\nColored by Quality Score', fontsize=16, pad=20)
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.grid(True, alpha=0.3)
plt.show()

n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
n_noise = list(cluster_labels).count(-1)
print(f"\nHDBSCAN found {n_clusters} stable clusters and {n_noise} outliers.")
```

## 11. Pros and cons of the approach [2,0 valor]

The WineQT dataset contains physicochemical properties of wines (such as acidity, pH, sulphates, alcohol) and a quality score. These two unsupervised techniques (Principal Component Analysis - PCA, and Self-Organizing Maps - SOM) were applied independently to explore the patterns and structure in this dataset. Each method offers its own advantages and limitations, when used separately.

1. **Pros of PCA:**

- **Dimensionality Reduction:** PCA compresses 11 correlated features into fewer uncorrelated components while preserving most variance. This simplifies analysis and reduces redundancy among acidity-related variables.
- **Variance Explanation:** PCA provides clear metrics (explained variance ratio) to identify which components capture the most information.
- **Visualization:** By projecting data onto the first two or three principal components, PCA enables intuitive scatter plots and biplots for pattern recognition.
- **Feature Insight:** Loadings reveal which physicochemical properties (e.g., alcohol, density) influence wine quality most strongly.

2. **Cons of PCA:**

- **Linear Assumption:** PCA only captures linear relationships. Non-linear interactions between features (common in wine chemistry) remain hidden.
- **Interpretability:** Principal components are abstract combinations of original features, making direct interpretation less intuitive for domain experts.
- **Outlier Sensitivity:** Extreme values can distort component directions.
- **Not a Clustering Tool:** PCA organizes data by variance, not by grouping similar wines into clusters.

3. **Pros of SOM:**

- **Non-linear Topology Preservation:** SOM maps high-dimensional data onto a 2D grid while maintaining neighborhood relationships, revealing complex patterns PCA might miss.
- **Clustering Capability:** SOM naturally groups wines with similar physicochemical profiles, which can correlate with quality levels.
- **Rich Visualization:** U-Matrix and component planes provide detailed insights into cluster boundaries and feature influence across the map.
- **Exploratory Power:** SOM uncovers hidden structures without requiring labels, useful for quality segmentation.

4. **Cons of SOM:**

- **Computational Cost:** SOM training is iterative and slower than PCA, especially for large grids or datasets.
- **Hyperparameter Sensitivity:** Grid size, learning rate, and neighborhood function require careful tuning for meaningful results.
- **Interpretation Complexity:** Understanding SOM outputs (U-Matrix, component planes) demands experience.
- **No Direct Variance Metric:** Unlike PCA, SOM does not quantify explained variance, making dimensionality reduction less explicit.

## 12. Future improvements [2,0 valor]

1. **Enhancements for PCA:**

- **Non-linear Dimensionality Reduction:** Traditional PCA captures only linear relationships. Future work could employ Kernel PCA, t-SNE, or UMAP to uncover non-linear structures in wine chemistry.
- **Dynamic Component Selection:** Instead of a fixed variance threshold (e.g., 95%), adaptive algorithms could determine the optimal number of components based on clustering performance or predictive accuracy.
- **Feature Selection Integration:** Combining PCA with feature selection methods (e.g., LASSO, mutual information) would improve interpretability by retaining the most influential physicochemical properties.
- **Robust PCA:** Implementing robust PCA variants can mitigate the impact of outliers, which often distort principal components.


2. **Enhancements for SOM:**

- **Hyperparameter Optimization:** SOM performance depends on grid size, learning rate, and neighborhood function. Automated tuning using Bayesian optimization or genetic algorithms could improve map quality.
- **Growing and Hierarchical SOMs:** Dynamic SOM architectures that expand or create hierarchical layers can better handle complex datasets and reduce quantization error.
- **Hybrid SOM Models:** Combining SOM with clustering algorithms (e.g., k-means on SOM codebooks) or supervised learning could enhance predictive capabilities for wine quality.
- **Interactive Visualization:** Advanced visualization tools, such as interactive component planes or 3D SOM grids, would make interpretation more intuitive for domain experts.
